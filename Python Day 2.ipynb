Count digits in a number

```python
def countDigits(n):
    cnt=0
    while n > 0:
        cnt = cnt + 1
        n = n // 10
    return cnt
if __name__ == "__main__":
     N=329823
     print("N:",N)
     digits=countDigits(N)
     print("Number of Digits in N:",digits)
```

Reverse the number

```python
class Solution:
    def reversenumber(self,n):
        revnum=0
        while n > 0:
            lastdigit=n%10
            revnum=revnum*10+lastdigit
            n=n//10
        return revnum
obj=Solution()
num=12345
print(obj.reversenumber(num))
```

Check if a number is Palindrome or Not

```python
def palindrome(n):
    revnum=0 #4
    dup=n
    while n>0:
        ld=n%10
        revnum=(revnum*10)+ld #4
        n//=10
    return dup ==revnum

number = 4554  # Example number
if palindrome(number):  # Check if the number is a palindrome
    print(f"{number} is a palindrome.")
else:
    print(f"{number} is not a palindrome.")
```
