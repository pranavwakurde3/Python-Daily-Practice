Adding the Two numbers using Class and Function

```python
def add_numbers(num1, num2):
  """
  This function takes two numbers as input and returns their sum.
  """
  sum_result = num1 + num2
  return sum_result

# Assign values to variables
number1 = 10
number2 = 20

# Call the function and store the result
result = add_numbers(number1, number2)

# Print the result
print(f"The sum of {number1} and {number2} is: {result}")
```

```python
class Adder:
    """
    This class adds two numbers using an instance method.
    """
    def __init__(self, num1, num2):
        self.num1 = num1  # Instance variable
        self.num2 = num2  # Instance variable

    def calculate_sum(self):
        """
        Calculates and returns the sum of the two numbers.
        """
        return self.num1 + self.num2

# Get user input for the numbers (converted to float)
number1_input = float(input("Enter first number: "))
number2_input = float(input("Enter second number: "))

# Create an instance (object) of the Adder class
# This automatically calls the __init__ method
adder_instance = Adder(number1_input, number2_input)

# Call the calculate_sum method on the instance to get the result
sum_result = adder_instance.calculate_sum()

# Display the sum
print(f"The sum of {number1_input} and {number2_input} is {sum_result}")
```

Print N to 1 using Recursion

```python
class Solution:
    def printNumber(self,current):
        if current < 1:
            return

        print(current,end=' ')

        self.printNumber(current - 1)

if __name__=='__main__':
  sol=Solution()
  n=10
    
  sol.printNumber(n)
  print()
```

```python
class Solution:
    # Recursive function to print numbers from current down to 1 using backtracking
    def printNumbers(self, current):
        # Base case: if current is less than 1, stop recursion
        if current < 1:
            return

        # Recursive call with previous number
        self.printNumbers(current - 1)

        # Print current number during backtracking
        print(current, end=' ')

if __name__ == "__main__":
    sol = Solution()
    n = 10

    sol.printNumbers(n)
    print()
```

Sum of first N Natural Numbers

```python
class Solution:
    # Function to find sum of first N natural numbers using loop
    def sumOfNaturalNumbers(self, N):
        # Initialize sum to 0
        total = 0
        
        # Loop from 1 to N
        for i in range (1, N + 1):
            # Add current number to sum
            total += i
        
        # Return the computed sum
        return total

# Driver code
if __name__ == "__main__":
    # Create object of Solution class
    obj = Solution()
    
    # Input value for N
    N = int(input())
    
    # Call the function and store the result
    result = obj.sumOfNaturalNumbers(N)
    
    # Print the result
    print(result)
```
