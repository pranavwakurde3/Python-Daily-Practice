**Factorial of a Number : Iterative and Recursive**

```python
def factorial(X):
    ans=1

    for i in range(1,X+1):
        ans*=i              #---> 5*1*2*3*4  =  120
    return ans

X=5

result = factorial(X)

print(f"The factorial of {X} is {result}")
```

**Reverse a given Array**

**1) Brute Force Apporach**

```python
class Solution:
    def reversearray(self,arr):
        n=len(arr)

        ans=[0]*n

        for i in range(n):
            ans[i] = arr [n-1-i]  # ---> 1) 5-1-0 = 4 ---> 2) 4-1-0 = 3 
        return ans

if __name__=='__main__':
    obj=Solution()
    arr=[1,2,3,4,5]
    result=obj.reversearray(arr)
    print("Reversed Array:",result)
```

**2) Better Approach**

```python
class Solution:
    def reversearray(self,arr):
        p1=0                # p1 = 0
        p2=len(arr) - 1     # p2 = 0 1 2 3 4 indices

        while p1 < p2:
            arr[p1],arr[p2]=arr[p2],arr[p1]  # ---> SWAP swap arr[0] and arr[4]

            p1+=1   #p1 = 1
            p2-=1   #p2 = 3  SWAP arr[1] and arr[3]
                                        
if __name__=='__main__':
    sol=Solution()
    arr=[1,2,3,4,5]
    sol.reversearray(arr)
    print(" ".join(map(str,arr)))
```

**3) Built in Library Function Approach**

```python
class Solution:
    def reversearray(self,arr):
        arr[:]=arr[::-1]  #array[start : end : step]

if __name__=='__main__':
    arr=[1,2,3,4,5]
    obj = Solution()
    obj.reversearray(arr)
    print(" ".join(map(str,arr)))
```
