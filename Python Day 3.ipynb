GCD

```python
def find_gcd(a,b):
    while a>0 and b>0: 
        if a>b:
            a=a%b
        else:
            b=b%a                   
    if a==0:
        return b
    return a
def main():
    n1=20
    n2=15

    gcd=find_gcd(n1,n2)
    print("GCD of", n1, "and", n2, "is:", gcd)


if __name__ == "__main__":
    main() 
```

Check if a number is Armstrong Number or not

```python
class ArmstrongChecker:
    def is_armstrong(num):
        k = len (str(num))
        sum = 0
        n = num

        while n>0:
            ld = n % 10            #153 % 10 --> 3   #15 % 10 --> 5  #1 % 10 --> 1
            sum += ld**k           # 3^3  + 5^3  +  1^3   --->  27+125=1 = 153
            n = n//10              #15  #1  #0
        return sum == num          #satisfied
        
number = 153
if ArmstrongChecker.is_armstrong (number):
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

Print all Divisors of a given Number

```python
class Solution:
    def getDivisors(self,n):
        res=[]

        for i in range(1, n+1):
           if n%i==0:
               res.append(i)
        return res

sol=Solution()
n=36
result=sol.getDivisors(n)
print("Divisors of",n, ":",*result)
```

Check if a number is prime or not

```python
def checkPrime(n):
    cnt = 0
    for i in range(1, n+1):
        if n % i == 0:
            cnt +=1
    return cnt == 2

n = 12
isPrime = checkPrime(n)
if isPrime:
    print(f"{n} is a prime number.")
else:
    print(f"{n} is not a prime number.")
```
